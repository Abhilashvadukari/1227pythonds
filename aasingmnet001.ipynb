```python
import numpy as np
import matplotlib.pyplot as plt


temps = np.array([25, 28, 30, 35, 40, 38, 36, 34, 32, 29, 26, 24])


norm_temps = (temps - temps.min()) / (temps.max() - temps.min())


avg_temp = temps.mean()
hottest = temps.max()
coldest = temps.min()


above_avg_months = np.where(temps > avg_temp)[0] + 1


temps_f = (temps * 9/5) + 32

print("Original Temps (C):", temps)
print("Normalized:", norm_temps)
print("Average:", avg_temp)
print("Hottest:", hottest, "Coldest:", coldest)
print("Months above average:", above_avg_months)
print("In Fahrenheit:", temps_f)


plt.plot(range(1,13), temps, marker='o', label="Celsius")
plt.axhline(avg_temp, color='r', linestyle='--', label="Average")
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.title("Monthly Temperature Trend")
plt.legend()
plt.show()
```
